#**Q2)What is the difference between Hard Margin and Soft Margin SVM?**
#ANS:-
* Hard margins are the marginal lines or plane which are made using the closest data point to the decision boundry. This makes the gap between them very small which can lead to overfitting and sometimes even wrong classification of data.

* Soft margins overcome the disadvantage of Hard margins by sacrificing some datapoints or by using some errors and we consider other data point as support vector to avoid overfitting and also get good accuracy ontest data.

```python
#Q24) Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary:
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0,n_clusters_per_class=1, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
model = SVC(kernel='poly', degree=3, C=1.0)
model.fit(X_train, y_train)
def plot_decision_boundary(model, X, y, title):
      x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1#Calculate ranges for plotting
      y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
      xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),np.linspace(y_min, y_max, 300))#We Plotting the gri region right now,
      grid = np.c_[xx.ravel(), yy.ravel()]#contains the gird points in 2d format
      Z = model.predict(grid).reshape(xx.shape)#is reshaped back to match the shape of the plot grid.
      plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.5)#Fills the background with different colors based on prediction (class 0 or 1).
      plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
      plt.title(title)
      plt.xlabel("Feature 1")
      plt.ylabel("Feature 2")
      plt.show()
plot_decision_boundary(model, X_test, y_test, "SVM with Polynomial Kernel (Test Set)")#plot_decision_boundary(model, X_test, y_test, "SVM with Polynomial Kernel (Test Set)")
```

```python
#Q27)Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually

C=[1,2,3]
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
for i, C in enumerate(C):
    model = SVC(kernel='linear', C=C)
    model.fit(X_train, y_train)

    # Create a mesh grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))

    # Predict on mesh
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot
    plt.subplot(1, 3, i+1)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolor='k')
    plt.title(f"C = {C}")
    plt.xticks([])
    plt.yticks([])

plt.tight_layout()
plt.show()

```

```python
#Q38) Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy
from sklearn.model_selection import StratifiedKFold
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Stratified K-Fold setup
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []

# Split and train manually using the folds
for train_idx, test_idx in skf.split(X, y):
    X_train = X[train_idx]
    X_test = X[test_idx]
    y_train = y[train_idx]
    y_test = y[test_idx]

    # Train SVM
    clf = SVC(kernel='linear')
    clf.fit(X_train, y_train)

    # Predict and calculate accuracy
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

# Print results
print("Fold Accuracies:", accuracies)
print("Average Accuracy:", np.mean(accuracies))
```

```python
#Q45)Write a Python program to train a NaÃ¯ve Bayes classifier and evaluate its performance using the ROC-AUC score
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Step 1: Generate a binary classification dataset
X, y = make_classification(n_samples=1000, n_features=20,
                           n_informative=10, n_redundant=5,
                           random_state=1)

# Step 2: Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=1)


model = GaussianNB()
model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)[:, 1]  # Probabilities for class 1

roc_auc = roc_auc_score(y_test, y_prob)
print(f"ROC-AUC Score: {roc_auc:.4f}")
fpr, tpr, _ = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr, label=f'NaÃ¯ve Bayes (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q46)Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt

# Step 1: Generate a binary classification dataset
X, y = make_classification(n_samples=1000, n_features=20,
                           n_informative=10, n_redundant=5,
                           random_state=42)

# Step 2: Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=42)

# Step 3: Train an SVM classifier with probability estimates
model = SVC(kernel='rbf', probability=True, random_state=42)
model.fit(X_train, y_train)

# Step 4: Get predicted probabilities
y_scores = model.predict_proba(X_test)[:, 1]

# Step 5: Compute precision and recall
precision, recall, _ = precision_recall_curve(y_test, y_scores)
avg_precision = average_precision_score(y_test, y_scores)

# Step 6: Plot Precision-Recall curve
plt.plot(recall, precision, label=f'AP = {avg_precision:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (SVM)')
plt.legend()
plt.grid(True)
plt.show()
```
